# Data Analysis 
Welcome! In this analysis, we are going to look at the dataset and see if we can see some trends and interesting points. 

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

# Load Dataset
Load the dataset into a pandas DataFrame.

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/melaniewalsh/responsible-datasets-in-context/main/datasets/top-500-novels/library_top_500.csv')

# Now, let's take a look at the dataset and see what we have here
df
```

There are a lot of interesting points to look at in the dataset! One helpful aspect to consider is the columns. We can get the list of all the columns in the dataset with the code below.

```python
df.columns
```

# Filter Relevant Columns
Filter the DataFrame to include only the 'pub_year' and 'genre' columns.

```python
# Filter the DataFrame to include only the 'pub_year' and 'genre' columns
df_filtered = df[['pub_year', 'genre']]

# Display the first few rows of the filtered DataFrame to ensure it filtered correctly
df_filtered.head()
```

# Group by Year and Genre
Now we can group the DataFrame by 'pub_year' and 'genre', and count the occurrences.

```python
# Group the DataFrame by 'pub_year' and 'genre', and count the occurrences
df_grouped = df_filtered.groupby(['pub_year', 'genre']).size().reset_index(name='count')

# Display rows of the grouped DataFrame to ensure it grouped correctly
df_grouped
```

```python
df_ratings = df.groupby(['gr_num_ratings_rank', 'gr_avg_rating']).size().reset_index(name='rating_count')
df_ratings
```

# Plot Genre Over Time
We can use altair to creat some visualization about the data and get a better understanding about the trends that are present.

```python
# Create a pie chart of the genre distribution
alt.Chart(df['genre'].value_counts().reset_index()).mark_arc().encode(
    theta='genre',
    color='index',
    tooltip=['index', 'genre']
).properties(
    title='Genre Distribution'
    
)
```

Great! We can see how each genre is distributed in the pie chart. "na" seems to be the most popular genre, with history coming in second, and fantasy in third. Although is it an amazing graph, one of the weaknesses of this type of chart is the lack of providing percise information. It is important to know what graph to use for each situation, so let's try something different.

```python
# Plot Number of Books per Genre
alt.Chart(df_grouped).mark_bar().encode(
	x="genre:N",
	y="count:Q",
	color="genre:N",
    
).properties(
	title="Number of books per Genre",
	width=600,
	height=600
)
```

This looks much better! Here we can see that this graph further proves the point that "na" is the most common genre, but it is able to provide us with the number of books in each genre, which in this case would be just over 200 books. 

Let's try to look further and see what we can find in the data. It would be interesting to see how these genres are presented over time. Luckily, we can make a bar graph to see this data.

```python
# Plot Genre Over Time
alt.Chart(df_grouped).mark_bar(size=10).encode(
    x='pub_year',
    y='count',
    color='genre',

).properties(
	title="Number of Books Published by Genre Over Time",
	width=600,
	height=600
)
```

Very interesting! This graph shows us two things. The first is that books were were not really recorded during the earler periods of time, which makes sense, as that data could have been lost over time. On the other hand, it is seen that in the modern age, there is a sharp increase in books being recorded and labled. The second is that "na" consistently shows up through time, even in the modern period where it is assumed that most books would be labled in another group that is not "na". 

This is all very fascinating! 

# Other Trends
Very good! Let's see if we can find other interesting trends on the data.

```python
# Create a dataframe with only the rows where the average rating is greater than or equal to 4.0
df_high_rank = df[(df["gr_avg_rating"] >= 4.0)]
df_high_rank
```

This dataframe shows only the books with an average rating of 4.0 or above. Let's create a graph with this to get a better sense of the data.

```python
# Create a bar chart with the average rating on the x-axis and the number of books on the y-axis
alt.Chart(df_high_rank).mark_bar().encode(
    x='gr_avg_rating',
    y='count()',
    color='genre'
).properties(
    title='Number of Books with Average Rating Greater Than or Equal to 4.0',
    width=600,
    height=600
)
```

Alright, very good! We can see that the data is skewed to the right, indicating that there are more books that have a lower rating than a high rating. There is also the interesting point that the most highly rated books are from the genre of "fantasy". 

Let's create a dataframe with the highest average ranking of books to see what the top rated books are!

```python
df_best = df_high_rank[(df_high_rank['gr_avg_rating'] >= 4.55) & (df_high_rank['genre'].notna())]
df_best
```

How interesting! In the top 5 highest rated books of all time, 4 of them are from one author, J.K. Rowling, and are from the same series, the Harry Potter series. I guess this means that I have to read the books some day since they are so highly rated.
